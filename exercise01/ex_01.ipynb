6-Given the digraph G = (V, E) where V = {M, N, O, P, Q, R, S} and
E ={(M, S), (N, O), (P, R), (N, S), (O, M), (N, Q), (O, M), (P, P), (S, M), (O, N),  (S, M), (N, R), (P, M), (M, S)}

(a) Specify, if any, a simple path from vertex M to vertex S.

(b) Specify, if any, a simple cycle, involving at least 4 nodes.

(c) Is the digraph connected or not connected?

(d) What is the degree of vertices N and R.

(e) Represent the digraph using adjacency list representation.

(f) Represent the digraph using adjacency matrix representation.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.DiGraph()
G.add_nodes_from(['M', 'N', 'O', 'P', 'Q', 'R', 'S'])
G.add_edges_from([('M', 'S'), ('N', 'O'), ('P', 'R'), ('N', 'S'), ('O', 'M'), ('N', 'Q'), ('O', 'M'), ('P', 'P'), ('S', 'M'), ('O', 'N'),  ('S', 'M'), ('N', 'R'), ('P', 'M'), ('M', 'S')])
```

```python
#Specify, if any, a simple path from vertex M to vertex S
nx.is_simple_path(G, ('M', 'S'))
```

```python
#Specify, if any, a simple cycle, involving at least 4 nodes
sorted(nx.simple_cycles(G))
```

```python
#Is the digraph connected or not connected?
nx.is_connected(G)
```

```python
nx.draw(G)
plt.show()
```

A função is_connected não é implementado para grafos direcionados, mas podemos ver na imagem que ele é conectado.

```python
#What is the degree of vertices N and R
nx.degree(G, ('N', 'R'))
```

```python
#Represent the digraph using adjacency list representation
for line in nx.generate_adjlist(G):
    print(line)
```

```python
#Represent the digraph using adjacency matrix representation
nx.adjacency_matrix(G).todense()
```

8-How many edges does a graph have 3 vertices of degree 3 and one vertex of degree 5?

9-Mr. A is friend with Mrs. B, but she doesn't like him back. She has a reciprocal friendship with both C and D, but only C considers D a friend.
```python
G = nx.DiGraph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('C', 'B')
G.add_edge('B', 'D')
G.add_edge('D', 'B')
G.add_edge('C', 'D')
G.add_edge('D', 'E')
G.add_edge('D', 'F')
G.add_edge('D', 'G')
G.add_edge('D', 'H')
G.add_edge('G', 'D')
G.add_edge('G', 'A')
G.add_edge('A', 'G')

nx.draw(G, with_labels=True)
plt.show()
```

10-Draw the graph from the previous exercise as undirected and weighted, with the weight being 2 if the connection is reciprocal, 1 otherwise.

```python
G = nx.Graph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

G.add_edge('A', 'B', weight=1)
G.add_edge('B', 'C', weight = 2)
G.add_edge('C', 'B', weight = 2)
G.add_edge('B', 'D', weight = 2)
G.add_edge('D', 'B', weight = 2)
G.add_edge('C', 'D', weight = 1)
G.add_edge('D', 'E', weight = 1)
G.add_edge('D', 'F', weight = 1)
G.add_edge('D', 'G', weight = 2)
G.add_edge('D', 'H', weight = 1)
G.add_edge('G', 'D', weight = 2)
G.add_edge('G', 'A', weight = 2)
G.add_edge('A', 'G', weight = 2)


pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.show()
```
